```python
# Quantum Harmonic Oscillator Approximation using PyTorch

import torch
import matplotlib.pyplot as plt

# Discretize space
x = torch.linspace(-5, 5, 1000)
dx = x[1] - x[0]

# Potential energy: V(x) = 0.5 * x^2 (harmonic oscillator)
V = 0.5 * x**2

# Construct Hamiltonian matrix (using finite differences)
diag = (1.0 / dx**2) + V
off_diag = -0.5 / dx**2 * torch.ones(x.shape[0] - 1)

H = torch.diag(diag) + torch.diag(off_diag, 1) + torch.diag(off_diag, -1)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = torch.linalg.eigh(H)

# Plot first 3 wavefunctions
plt.figure(figsize=(10, 5))
for i in range(3):
    plt.plot(x, eigenvectors[:, i], label=f'n={i}')
plt.title("Quantum Harmonic Oscillator Eigenstates")
plt.xlabel("Position")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
